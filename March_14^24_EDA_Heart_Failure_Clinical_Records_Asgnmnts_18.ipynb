```python
#5.Is there any correlation between creatinine and sodium level in the blood?
```

```python
df.columns
```

```python
df[["serum_creatinine","serum_sodium"]]
```

```python
#Since both creatinine and sodium levels are numeric data types, Using scatter plot to visualise:

#Ploting serum_creatinine vs serum_sodium:
plt.scatter(df["serum_sodium"],df["serum_creatinine"])

#Setting lebels
plt.xlabel("Creatinine_level")
plt.ylabel("Sodium_level")
plt.title("Creatinine vs Sodium level")

#Displaying the plot
plt.show()
```

```python
correlation = np.corrcoef(df["serum_sodium"], df["serum_creatinine"])[0, 1]

# Print the correlation coefficient
print("The correlation coefficient between serum_sodium and serum_creatinine is:", correlation)
```

```python
'''Insight: Based on the correlation coefficient of -0.189, we can conclude that there is a weak negative correlation between serum sodium and
serum creatinine levels in the blood. However, it's important to note that correlation does not imply causation, and there may be other factors at play.
Further analysis and consideration of additional variables are recommended to gain a more comprehensive understanding of the relationship.'''
```

```python
#9.Is there any significant difference in ejection fraction between patient with and without diebetes?
```

```python
df
```

```python
df.columns
```

```python
df[["diabetes","ejection_fraction"]].value_counts(normalize=True)
```

```python
diabetic_patients = df[df['diabetes'] == 1]
non_diabetic_patients = df[df['diabetes'] == 0]
```

```python
diabetic_patients['ejection_fraction']
```

```python
non_diabetic_patients['ejection_fraction']
```

```python

plt.subplot(1,2,1)
sns.histplot(diabetic_patients['ejection_fraction'],kde=True,color="skyblue")
plt.title("Distributin for Dibetic patient")
plt.xlabel("Ejection Fraction")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
sns.histplot(non_diabetic_patients['ejection_fraction'],kde=True,color="skyblue")
plt.title("Distributin for non Diabetic patient")
plt.xlabel("Ejection Fraction")
plt.ylabel("Frequency")

plt.show()
```

```python
mean_ejection_fraction_diabetic=diabetic_patients['ejection_fraction'].mean()
print("Mean Ejection Fraction of people with diabetes:",mean_ejection_fraction_diabetic)
```

```python
mean_ejection_fraction_non_diabetic=non_diabetic_patients['ejection_fraction'].mean()
print("Mean Ejection Fraction of people without diabetes:",mean_ejection_fraction_non_diabetic)
```

```python
difference_bw_ejection_fraction = mean_ejection_fraction_non_diabetic-mean_ejection_fraction_diabetic
print("Difference in ejection fraction between patient with and without diebetes: ",difference_bw_ejection_fraction)
```

```python
"""Insight:Based on the calculated difference in ejection fraction between patients with and without diabetes, it seems that there is a slight variation
in ejection fraction based on the presence of diabetes. However, it's important to consider other factors that may influence high blood pressure as well.
Further analysis and investigation would be needed to draw a conclusive understanding of the relationship between diabetes and ejection fraction."""
```

```python
survived["serum_creatinine"]
```

```python
Not_survived["serum_creatinine"]
```

```python
plt.subplot(1,2,1)
sns.histplot(survived['serum_creatinine'],kde=True,color="skyblue",bins=10)
plt.title("Creatinine level in Survived")
plt.xlabel("Creatinine level")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
sns.histplot(non_diabetic_patients['serum_creatinine'],kde=True,color="skyblue",bins=10)
plt.title("Creatinine level in not survived")
plt.xlabel("Creatinine")
plt.ylabel("Frequency")

plt.show()
```

```python
mean_cr_level_Surv=survived["serum_creatinine"].mean()
mean_cr_level_not_surv=Not_survived["serum_creatinine"].mean()
Difference_between_Survived_and_Not_survived=mean_cr_level_not_surv-mean_cr_level_Surv
print("Difference in creatinine level between those who survived and those who didn't: ",Difference_between_Survived_and_Not_survived)
```
